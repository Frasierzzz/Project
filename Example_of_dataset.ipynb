# scitldr dataset (Extreme Summarization of Scientific Documents)



มีข้อมูล 5.4k TLDR จากเอกสาร 3.2k ฉบับ

TLDR หมายถึง Too Long; Didn't Read แปลว่า ยาวไปไม่อ่าน! ในที่นี้ก็คือ "ประโยคสั้นๆ"

| | train | valid | test |
|----------|----------|----------|-----|
| SciTLDR-A	   | 1992  | 618  | 619  |
| SciTLDR-AIC	   | 1992  | 618  | 619  |
| SciTLDR-FullText	   | 1992  | 618  | 619  |

โครงสร้างหลัก คือ
**source :** ข้อความ หรือเนื้อหาที่ต้องการให้โมเดลสรุป

**source_labels :** บอกว่าข้อความส่วนไหนของ source มีความสำคัญ เป็นลิสต์ที่มีค่า True/False สำหรับแต่ละประโยคใน source

**rouge_scores :** คำนวณจากการเปรียบเทียบข้อความต้นฉบับกับ target summary ใช้ประเมินว่า summary ที่สร้างได้ตรงกับความสำคัญในต้นฉบับมากแค่ไหน
ใช้ ROUGE-1

**paper_id :** รหัสที่ใช้ระบุบทความ อิงจาก arXiv

**target :** ข้อความสรุป aka TLDR

โดยแบ่งเป็น 2 ประเภท คือ Author-written TLDRs (จากผู้เขียน) และ Expert-derived TLDRs (จากผู้เชี่ยวชาญ)

**ปล.**
ROUGE (Recall-Oriented Understudy for Gisting Evaluation) เป็นชุดตัวชี้วัดที่ใช้ประเมินคุณภาพของข้อความที่ได้จากการสรุป (หรือการสร้างข้อความใหม่) โดยเปรียบเทียบข้อความที่ได้กับข้อความที่เป็น "gold standard" หรือข้อความต้นฉบับที่มนุษย์สร้างขึ้น

โดย ROUGE แบ่งเป็น Recall, Precision, F1-Score ซึ่ง score ที่อยู่ในข้อมูลคือ F1-Score

Recall = จำนวนคำที่ตรง / จำนวนคำใน target

Precision = จำนวนคำที่ตรง / จำนวนคำในข้อความ หรือบทความนั้น

F1-Score (ROUGE-1) = 2⋅ (Precision⋅Recall)  / (Precision+Recall)
